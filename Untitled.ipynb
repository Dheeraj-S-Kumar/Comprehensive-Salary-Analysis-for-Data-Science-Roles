# finding outliers

```python
def wisker(col):
    q1, q3 = np.percentile(col, [25, 75])
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return lower_bound, upper_bound
```

```python
mean_salary_by_year = df.groupby('work_year')['salary_in_usd'].mean()

mean_salary_by_year.plot(kind='bar')
plt.title('Mean Salary by Work Year')
plt.xlabel('Work Year')
plt.ylabel('Mean Salary (USD)')
plt.xticks(rotation=45) 
plt.show()
```

```python
plt.figure(figsize=(40, 20))
sb.boxplot(x='job_title', y='salary_in_usd', data=df)
plt.title('Distribution of Salary by Job Title')
plt.xlabel('Job Title')
plt.ylabel('Salary (USD)')
plt.xticks(rotation=90)  
plt.tight_layout()  
plt.show()
```

```python
mean_salary_by_year = df.groupby('experience_level')['salary_in_usd'].mean()

mean_salary_by_year.plot(kind='bar')
plt.title('Mean Salary by experience_level')
plt.xlabel('experience_level')
plt.ylabel('Mean Salary (USD)')
plt.xticks() 
plt.show()
```

```python
mean_salary_by_year = df.groupby('employment_type')['salary_in_usd'].mean()

mean_salary_by_year.plot(kind='bar')
plt.title('Mean Salary by employment_type')
plt.xlabel('employment_type')
plt.ylabel('Mean Salary (USD)')
plt.xticks(rotation=45) 
plt.show()
```
